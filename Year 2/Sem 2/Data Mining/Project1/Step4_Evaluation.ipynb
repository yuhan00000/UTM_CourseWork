```python
# Step 4: Evaluation and Comparison Across Models
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# ========== Input Model Scores ========== #

# Insight 1: Attrition
insight1 = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest', 'KNN'],
    'Accuracy': [87.76, 83.67, 84.58],
    'Precision': [70.73, 47.37, 61.54],
    'Recall': [40.85, 12.68, 11.27],
    'F1 Score': [51.79, 20.00, 19.05]
})

# Insight 2: Performance Rating (macro-average)
insight2 = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest', 'KNN'],
    'Accuracy': [97.73, 100.00, 83.45],
    'Precision': [97.33, 100.00, 53.47],
    'Recall': [93.85, 100.00, 50.53],
    'F1 Score': [95.49, 100.00, 48.06]
})

# Insight 3: Job Hopper
insight3 = pd.DataFrame({
    'Model': ['Logistic Regression', 'Random Forest', 'KNN'],
    'Accuracy': [99.55, 100.00, 72.56],
    'Precision': [99.45, 100.00, 71.03],
    'Recall': [99.45, 100.00, 56.59],
    'F1 Score': [99.45, 100.00, 63.00]
})
```

```python
# ========== Function to Plot Bar Chart ========== #
def plot_comparison_table(df, title):
    print("\n" + title + "\n")
    display(df.round(2))
    
    metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
    df_melted = df.melt(id_vars='Model', value_vars=metrics, var_name='Metric', value_name='Score')
    
    plt.figure(figsize=(10, 6))
    sns.barplot(data=df_melted, x='Metric', y='Score', hue='Model')
    plt.title(title)
    plt.ylabel("Score (%)")
    plt.ylim(0, 110)
    plt.legend(title='Model')
    plt.grid(True, axis='y')
    plt.show()
```

```python
# ========== Display Tables and Charts ========== #
plot_comparison_table(insight1, "Insight 1: Predicting Attrition")
plot_comparison_table(insight2, "Insight 2: Predicting Performance Rating")
plot_comparison_table(insight3, "Insight 3: Predicting Job Hopper")
```

```python
# ========== Line Chart Across All Models (Per Metric) ========== #
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']

for i, (df, title) in enumerate(zip([insight1, insight2, insight3], 
                                    ["Insight 1: Attrition", "Insight 2: Performance Rating", "Insight 3: Job Hopper"])):
    plt.figure(figsize=(10, 6))
    for j in range(len(df)):
        plt.plot(metrics, df.iloc[j, 1:], marker='o', label=df['Model'][j])
    plt.title(f"Metric Trends Across Models - {title}")
    plt.xlabel("Metrics")
    plt.ylabel("Score (%)")
    plt.ylim(0, 110)
    plt.legend()
    plt.grid(True)
    plt.show()
```
