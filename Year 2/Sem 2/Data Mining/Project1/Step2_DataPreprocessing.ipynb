```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
# Make sure the Excel file is in the same folder as your notebook
df = pd.read_excel("employee dataset.xlsx")

# Show first 5 rows
df.head()
```

```python
# Identify columns with only one unique value
constant_cols = [col for col in df.columns if df[col].nunique() == 1]
print("Constant columns to drop:", constant_cols)

# Drop constant and irrelevant columns
df.drop(columns=['EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours'], inplace=True)
```

```python
# Convert 'Attrition' into binary values
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
# Label encode 'Gender' and 'OverTime' as they have 2 categories
binary_cols = ['Gender', 'OverTime']
le = LabelEncoder()

for col in binary_cols:
    df[col] = le.fit_transform(df[col])
```

```python
# One-hot encode categorical columns with more than 2 values
multi_cat_cols = ['BusinessTravel', 'Department', 'EducationField', 'JobRole', 'MaritalStatus']
df = pd.get_dummies(df, columns=multi_cat_cols)
```

```python
# Add binary target for Insight 3
df['JobHopper'] = np.where(df['NumCompaniesWorked'] >= 3, 1, 0)
```

```python
# Make sure there are no missing values
print("Missing values per column:\n", df.isnull().sum())
```

```python
# Save cleaned data to CSV for future use
df.to_csv("cleaned_employee_data.csv", index=False)
print("Cleaned dataset saved as cleaned_employee_data.csv")
```
