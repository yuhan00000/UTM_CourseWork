```python
# 01_preprocessing.ipynb

# 1. DATA LOADING

import pandas as pd
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler

# Load the dataset (make sure the file is in the same directory)
df = pd.read_excel("marketing_campaign.xlsx")

# Preview the data
print("Initial data shape:", df.shape)
df.head()
```

```python
# 2. DATA CLEANING

# Drop irrelevant columns: identifiers, campaign responses, and unused features
columns_to_drop = [
    'ID', 'Z_CostContact', 'Z_Revenue', 'Response',
    'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3',
    'AcceptedCmp4', 'AcceptedCmp5'
]
df.drop(columns=columns_to_drop, inplace=True)

# Drop rows with missing values in critical fields (only 'Income' has NaNs)
df.dropna(subset=['Income'], inplace=True)
```

```python
# 3. DATA TRANSFORMATION

# Create a new feature: Age
df['Age'] = 2025 - df['Year_Birth']

# Create a new feature: Customer_Tenure (in days)
df['Customer_Tenure'] = (datetime(2025, 6, 20) - df['Dt_Customer']).dt.days

# Drop original date columns and birth year since they are now transformed
df.drop(columns=['Dt_Customer', 'Year_Birth'], inplace=True)

# One-hot encode categorical features
# - Drop first category to avoid dummy variable trap
df = pd.get_dummies(df, columns=['Education', 'Marital_Status'], drop_first=True)
```

```python
# 4. DATA STANDARDIZATION (Normalization)

# Normalize all numerical features using Min-Max Scaling
scaler = MinMaxScaler()
df[df.columns] = scaler.fit_transform(df[df.columns])

# Final data preview
print("✅ Preprocessing complete. Final dataset shape:", df.shape)
df.head()
```

```python
# 5. SAVE FOR FUTURE USE

df.to_csv("preprocessed_customer_data.csv", index=False)
print("📁 Preprocessed data saved as 'preprocessed_customer_data.csv'")
```
